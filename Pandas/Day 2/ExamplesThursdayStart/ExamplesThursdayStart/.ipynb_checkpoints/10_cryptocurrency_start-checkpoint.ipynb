```python
import pandas as pd
import os
```

```python
bitcoin_csv = os.path.join("Resources", "bitcoin_cash_price.csv")
dash_csv = os.path.join("Resources", "dash_price.csv")
```

```python
bitcoin_df = pd.read_csv(bitcoin_csv)
dash_df = pd.read_csv(dash_csv)
```

```python
bitcoin_df.head()
```

```python
dash_df.head()
```

```python
# Merge the two DataFrames together based on the Dates they share
merge_table = pd.merge(bitcoin_df, dash_df, on="Date", how="inner")
merge_table.head()
```

```python
# Collecting best open for Bitcoin and Dash (Hint: .max())
btc_open = merge_table["Bitcoin Open"].max()
dash_open = merge_table["Dash Open"].max()
```

```python
# Collecting best close for Bitcoin and Dash
btc_close = merge_table["Bitcoin Close"].max()
dash_close = merge_table["Dash Close"].max()
```

```python
# Collecting the total volume for Bitcoin and Dash (in millions -- /1000000)
btc_total_vol = merge_table["Bitcoin Volume"].sum()/1000000
dash_total_vol = merge_table["Dash Volume"].sum()/1000000
```

```python
# Create a summary DataFrame using above values
summary_df = pd.DataFrame({"Best Bitcoin Open": [btc_open],
                           "Best Bitcoin Close": [btc_close],
                           "Total Bitcoin Valume": str([btc_total_vol]),
                           "Best Dash Open": [dash_open],
                           "Best Dash Close": [dash_close],
                           "Total Dash Volume": str([dash_total_vol])})
summary_df
```
