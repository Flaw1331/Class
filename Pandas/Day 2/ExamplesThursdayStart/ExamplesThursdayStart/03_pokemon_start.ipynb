```python
import pandas as pd
import os
```

```python
data_file = os.path.join("Resources", "Pokemon.csv")
pokemon_pd = pd.read_csv(data_file)
pokemon_pd.head()
```

```python
# Create a new DataFrame by extracting the following columns: 
#     Type 1, HP, Attack, Sp. Atk, Sp. Def, and Speed.
pokemon_type = pokemon_pd[["Type 1", "HP", "Attack", "Defense", "Sp. Atk", "Sp. Def", "Speed"]]
pokemon_type.head()
```

```python
# Group pokemons by 'Type 1'
pokemon_group = pokemon_type.groupby('Type 1')
pokemon_group.describe().head()
```

```python
# Create a new DataFrame with the mean values grouped by Type 1
group_pokemon_df = pokemon_type.groupby('Type 1').mean()
group_pokemon_df.head()
```

```python
# Add a new column with the total points per Type 1.
group_pokemon_df["Total"] = group_pokemon_df["HP"] + group_pokemon_df["Attack"] + group_pokemon_df["Defense"] + group_pokemon_df["Sp. Atk"] + group_pokemon_df["Sp. Def"] + group_pokemon_df["Speed"]
group_pokemon_df.head()
```
