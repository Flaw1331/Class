```python
# Import the Pandas library
import pandas as pd
import os
```

```python
# Create a reference the CSV file desired
data_file = os.path.join("Resources", "ufoSightings.csv")
# Read the CSV into a Pandas DataFrame
ufo_df = pd.read_csv(data_file)
# Print the first five rows of data to the screen
ufo_df.head()
```

```python
ufo_df.dtypes
```

```python
ufo_df["duration (seconds)"] = pd.to_numeric(ufo_df["duration (seconds)"],errors='coerce')
```

```python
ufo_df["duration (seconds)"][27820:27824]
```

```python
ufo_df.iloc[:,9] = pd.to_numeric(ufo_df.iloc[:,9], errors='coerce')
```

```python
ufo_df.dtypes
```

```python
ufo_df["duration (seconds)"].sum()
```

```python
ufo_df.count()
```

```python
ufo_df["state"] = ufo_df["state"].fillna("NR")
```

```python
ufo_df.head()
```

```python
clean_ufo_df = ufo_df.dropna(how="any")
```

```python
clean_ufo_df.count()
```

```python
usa_ufo_df = clean_ufo_df.loc[clean_ufo_df['country'] == "us", :]
usa_ufo_df.head()
```

```python
state_counts = usa_ufo_df["state"].value_counts()
```

```python
state_ufo_counts_df = pd.DataFrame(state_counts)
state_ufo_counts_df.head()
```

```python
state_ufo_counts_df.columns = ["Sum of Sightings"]
state_ufo_counts_df.head()
```

```python
test = usa_ufo_df.groupby('state')
type(test)
```

```python
group_usa_df = usa_ufo_df.groupby('state')["duration (seconds)"].sum()
group_usa_df.head()
```

```python
state_summary_table = pd.DataFrame({"Number of Sightings": state_counts,
                                   "Total Visit Time": group_usa_df})
state_summary_table.head()
```
