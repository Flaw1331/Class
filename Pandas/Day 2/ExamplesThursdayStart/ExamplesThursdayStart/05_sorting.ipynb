```python
# Import Dependencies
import pandas as pd
import os
```

```python
data_file = os.path.join("Resources", "Happiness_2017.csv")
happiness_df = pd.read_csv(data_file)
happiness_df.head()
```

```python
happiness_df.rename(columns={'Economy..GDP.per.Capita.': 'GDP'}, inplace=True)
happiness_df.head()
```

```python
gdp_df = happiness_df.sort_values("GDP",ascending=False)
gdp_df.head()
```

```python
freedom_df = happiness_df.sort_values(["GDP", "Freedom"],ascending=False)
freedom_df.head()
```
