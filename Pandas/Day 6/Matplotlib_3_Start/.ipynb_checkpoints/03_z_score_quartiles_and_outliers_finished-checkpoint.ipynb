```python
import matplotlib.pyplot as plt
from scipy.stats import zscore
import pandas as pd

data_frame = pd.DataFrame({'numbers':[8, 8, 12, 24, 54, 54, 75, 78, 98, 102, 132, -110]})
data_frame
```

```python
# With the data_frame above, find the lower and upper quartiles
lower = data_frame['numbers'].quantile(.25)
upper = data_frame['numbers'].quantile(.75)
```

```python
# add a column to the dataframe labelled z-score
# and compute the z-score for each item.
# since I'm considering this sample 'the population' i'll set ddof=0
data_frame['zscore'] = zscore(data_frame['numbers'], ddof=0)
data_frame
```

```python
# create box and whisker plot of the numbers column
import seaborn as sns
sns.set_style("whitegrid")
ax = sns.boxplot(y=data_frame['numbers'])
ax = sns.swarmplot(y=data_frame['numbers'], color='k')
plt.show()
```

```python
# Calculate the Interquartile Range. IQR = Q3 - Q1
iqr = upper - lower
print(iqr)
```

```python
iqr_15 = iqr * 1.5
```

```python
data_frame['is_outlier'] = -data_frame['numbers'].between(lower - iqr_15, upper + iqr_15)
```

```python
# Remove outliers from the data set.
```

```python
data_frame
```

```python
data_frame = data_frame[data_frame['is_outlier'] != True]
```

```python
data_frame.drop(data_frame.index[[0,1]], axis=0, inplace=True)
data_frame
```
