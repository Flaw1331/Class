## Correlation Matrix

The correlation matrix measures linear dependence between pairs of features. 

Mathematically, it uses the **Pearson Product-Moment Correlation Coefficient**, also known as **Pearson's r**. 

* R = 1. Two features have a perfect positive correlation
* R = 0. No correlation between two feature
* R = -1. Two features have a perfect negative correlation 

If you open a college Statistics textbook you'll see the complicated formula for setting this up. With Pandas and Seaborn we're going to do this automatically

```python
import seaborn as sns
attributes = ['sqft_living', 'zipcode','yr_built', 'bedrooms',
              'bathrooms', 'floors', 'waterfront', 'view', 
              'yr_renovated', 'lat', 'long', 'sqft_lot', 'price']
cm = np.corrcoef(housing_data[attributes].values.T)
fig, ax = plt.subplots(figsize=(10,10))
sns.heatmap(cm, cbar=True, annot=True, square=True,
            fmt='.2f', annot_kws={'size':15}, yticklabels=attributes, xticklabels=attributes)
plt.tight_layout()
```

```python
X = housing_data['sqft_living']
X = X.values.reshape(-1,1) # Sci-kit learn wants x values as a multi-dimensional array, not just sqft_living
Y = housing_data['price']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.3, random_state=2)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
def RegressionPlot(X, Y, model):
    plt.figure(figsize=(8,8))
    plt.scatter(X, Y, c='steelblue', edgecolor='white', s=70)
    plt.plot(X, model.predict(X), color='black', lw=2)
    return 

def ResidualPlot(X_train, X_test, y_train, y_test, model):
    y_train_pred = model.predict(X_train)
    y_test_pred = model.predict(X_test)
    plt.figure(figsize=(10,10))
    plt.scatter(y_train_pred,  y_train_pred - y_train,
            c='steelblue', marker='o', edgecolor='white',
            label='Training data', s=30)
    plt.scatter(y_test_pred,  y_test_pred - y_test,
            c='limegreen', marker='s', edgecolor='white',
            label='Test data', s=30)
    plt.xlabel('Predicted values')
    plt.ylabel('Residuals')
    plt.legend(loc='upper left')
    plt.hlines(y=0, xmin=-10, xmax=2000000, color='black', lw=2)
    plt.tight_layout()
    return

def TrueValue(X_test, y_test, model):
    fig = plt.figure(figsize=(8,8))
    plt.scatter(y_test, model.predict(X_test), alpha = 0.8, s=2)
    plt.plot(y_test, y_test, color='black', lw=1)
    plt.xlabel('True Test values Price ($)')
    plt.ylabel('My Estimation Price ($)')
```

```python
RegressionPlot(X, Y, model)  
plt.xlabel('Square Footage')
plt.ylabel('Price')
```

## Coefficient of Determination ##

From Wikipedia:
In statistics, the coefficient of determination, denoted R2 or r2 and pronounced "R squared", is the proportion of the variance in the dependent variable that is predictable from the independent variable(s).

It is a statistic used in the context of statistical models whose main purpose is either the prediction of future outcomes or the testing of hypotheses, on the basis of other related information. It provides a measure of how well observed outcomes are replicated by the model, based on the proportion of total variation of outcomes explained by the model.

* Best possible score is 1.0. 

* Sci-kit learn's default score function gives you R^2, coefficient of determination values.

* For our first linear regression model, 48.3% of the variability of Y can be explained using X.


```python
model.score(X_test, y_test)
```

```python
print('Slope: %.3f' % model.coef_[0])
print('Intercept: %.3f' % model.intercept_)
# for every increase in square feet we get $224.962
```

```python
TrueValue(X_test, y_test, model)
```

```python
ResidualPlot(X_train, X_test, y_train, y_test, model)
```

## Multiple linear regression

Of course, we can take linear regression in one dimension and push it out to multiple dimensions. Now, instead of just looking at square feet, we can include other variables.

This equation:

![equation](regression_eq.png)

The problem is, this is very hard to visualize. In two dimensions, you can think of a 'hyperplane', where the square foot regression is happening alongside another regression line to form a plane.

![hyperplane](hyperplane.png)

Of course, in the example below I'm using 12 attributes, so it's no so easy to think about what this plane looks like in 12 dimensions!

```python
attributes = ['sqft_living', 'zipcode','yr_built', 'bedrooms',
              'bathrooms', 'floors', 'waterfront', 'view', 
              'yr_renovated', 'lat', 'long', 'sqft_lot']
X = housing_data[attributes]
Y = housing_data['price']
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.3, random_state=12)
model2 = LinearRegression()
model2.fit(X_train, y_train)
```

```python
model2.score(X_train, y_train)
```

```python
TrueValue(X_test, y_test, model2)
```

```python
ResidualPlot(X_train, X_test, y_train, y_test, model2)
```

To be honest, I'm not being completely fair to multiple linear regression. Since there are multiple methods and ways we could improve this. I could normalize the data set, which would help. But, in my opinion, I'd rather jump to Gradient Boosting.

** GradientBoostingRegressor **

Though there are many possible supervised learning model types to choose from, gradient boosted models (GBMs) are almost always my first choice. Compared to Random Forests, they are faster. In many cases, they end up outperforming other options, and even when they don't, it's rare that a properly tuned GBM is far behind the best model.

At a high level, the way GBMs work is by starting with a rough prediction and then building a series of decision trees, with each tree in the series trying to correct the prediction error of the tree before it.

There are a ton of models in sci-kit learn. If there's one that shows up the most in winning competitions, it's Gradient Boosting Decision Trees. Netflix Prize for example was won this way.

For more information:
* https://www.youtube.com/watch?v=-5l3g91NZfQ&index=1&list=PLE5ASvOuOkqvMqkecblqJ79-Px5hwda07
* https://github.com/pprett/pydata-gbrt-tutorial


```python
from sklearn.ensemble import GradientBoostingRegressor

X = housing_data[['sqft_living']].values
y = housing_data['price'].values

model = GradientBoostingRegressor(random_state=12, n_estimators=10, max_depth=4)
model.fit(X, y)

sort_idx = X.flatten().argsort()
RegressionPlot(X[sort_idx], y[sort_idx], model)
plt.show()
```

```python
model.score(X, y)
```

Instead of a straight line regression we get a step-wise function.

![trees](trees.png)

>Individual decision trees can be interpreted easily by simply visualizing the tree structure. Gradient boosting models, however, comprise hundreds of regression trees thus they cannot be easily interpreted by visual inspection of the individual trees.  - http://scikit-learn.org/dev/modules/ensemble.html#partial-dependence

This is much harder to visualize than linear regression. In two dimensions, someone has done a great javascript visualization we can look at: 

http://arogozhnikov.github.io/2016/06/24/gradient_boosting_explained.html

```python
## Gradient Boosting Regressor won't take a datetime as a feature
# So let's get the days from the start.
# this feature should show some of the increases in housing prices for the year
housing_data.date.describe()
housing_data['datedays'] = (housing_data.date - min(housing_data.date)).astype('timedelta64[D]')
del housing_data['date']
```

```python
housing_data.columns
```

```python
from sklearn.ensemble import GradientBoostingRegressor
attributes = ['bedrooms', 'bathrooms', 'sqft_living', 'sqft_lot',
       'floors', 'waterfront', 'view', 'condition', 'grade', 'sqft_above',
       'sqft_basement', 'yr_built', 'yr_renovated', 'zipcode', 'lat', 'long', 'datedays']
X = housing_data[attributes]
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.3, random_state=12)
model = GradientBoostingRegressor(n_estimators=3000, max_depth=6, learning_rate=0.01,
                                loss='huber', random_state=0, verbose=1)
model.fit(X_train,y_train)
```

```python
model.score(X_test,y_test) 
```

```python
model.score(X_train,y_train)
```

```python
TrueValue(X_train, y_train, model)
```

```python
TrueValue(X_test, y_test, model)
```

```python
ResidualPlot(X_train, X_test, y_train, y_test, model)
```

To be honest, there are some improvements I could do... Scaling, feature engineering and feature transformation, parameter tuning. But this is a good start.

In linear regression, it was very easy to determine what was going on. With Boosting, as we've seen what is happening is much more opaque.


# Partial dependence

Partial dependence plots show the effect of a set of features, marginalizing over the values of all other features. If no other feature matters, the plot shows how this feature affects the housing price. 

The plot below contains one-way partial dependence plots (PDP) each showing the effect of an idividual feature on the housing price. 

We can see that sqft_living has a closely linear relationship with the house value. Latitude affects housing prices. As we get close to 47.6, where Seattle is located. It dips again past 47.7, which is lower King County near Tacoma.


```python

# sort importances
#for feature, importances in zip(housing_data.columns, model.feature_importances_)
#attributes = ['bedrooms', 'bathrooms', 'sqft_living', 'sqft_lot',
#       'floors', 'waterfront', 'view', 'condition', 'grade', 'sqft_above',
#       'sqft_basement', 'yr_built', 'yr_renovated', 'zipcode', 'lat', 'long',
#       'sqft_living15', 'sqft_lot15', 'datedays']
from sklearn.ensemble.partial_dependence import plot_partial_dependence
features = [1, 2, 14, (14,2), (15,14)]
fig, axs = plot_partial_dependence(model, X_train, features,feature_names=attributes, 
                                   figsize=(12, 8))
```
