```python
import pandas as pd
import numpy as np
import os
from scipy.stats import sem
from matplotlib import pyplot as plt
```

```python
filename = os.path.join("Resources","housing_data.csv")

housing_data = pd.read_csv(filename, header=None)
#housing_data = housing_data.sample(frac=1).reset_index(drop=True)
```

```python
housing_data
```

```python
div = 100
samples = []
for i in range(0, 10):
    samples.append(housing_data.take(np.random.permutation(len(housing_data))[:div]))

```

```python
# Calculate means
means = [np.mean(s)[0] for s in samples]
# Calculate standard error of the mean
standard_errors = [sem(s)[0] for s in samples]

```

```python
fig, ax = plt.subplots()

ax.errorbar(np.arange(0, len(means)), means, yerr=standard_errors, fmt="o", color="b",
            alpha=0.5, label="Mean of House Prices")

ax.set_xlim(-0.5, len(means))

ax.set_xlabel("Sample Number")
ax.set_ylabel("Mean of Median House Prices")

plt.legend(loc="best", fontsize="small", fancybox=True)
plt.show()
```
