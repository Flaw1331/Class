```python
# Import Dependencies
import random
import matplotlib.pyplot as plt
import numpy as np
```

```python
# The maximum x value for our chart will be 100
x_limit = 100

# List of values from 0 to 100 each value being 1 greater than the last
x_axis = np.arange(0, x_limit, 1)

# Create a random array of data that we will use for our y values
data = [random.random() for value in x_axis]
```

```python
# Tells matplotlib that we want to make a scatter plot
# The size of each point on our plot is determined by their x value
plt.scatter(x_axis, data, marker="o", facecolors="red", edgecolors="black",
            s=x_axis, alpha=0.75)
```

```python
# The y limits of our scatter plot is 0 to 1
plt.ylim(0, 1)
```

```python
# The x limits of our scatter plot is 0 to 100
plt.xlim(0, x_limit)
```

```python
# Prints the scatter plot to the screen
plt.show()
```

# Figure Instance

See https://matplotlib.org/api/_as_gen/matplotlib.figure.Figure.html#matplotlib.figure.Figure

Example:
* figure(figsize=(1, 1), dpi=72, facecolor='r')

* The default dpi is 72 (dots per inch)
* So figure(figsize=(1,1)) creates an inch by inch image, which will be 72-by-72 pixels unless you also give a different dpi argument.

```python
plt.figure(figsize=(1,1), facecolor='red')
plt.scatter(x_axis, data, marker="o", facecolors="red", edgecolors="black",
            s=x_axis, alpha=0.75)
plt.show()
```

```python
# technically This makes the figure's width 10 inches, and its height 10 inches.
plt.figure(figsize=(10,10))
plt.scatter(x_axis, data, marker="o", facecolors="red", edgecolors="black",
            s=x_axis, alpha=0.75)
plt.show()
```

# Tight Layout
tight_layout automatically adjusts subplot params so that the subplot(s) fits in to the figure area. This is an experimental feature and may not work for some cases. It only checks the extents of ticklabels, axis labels, and titles.

```python
plt.tight_layout()
plt.scatter(x_axis, data, marker="o", facecolors="red", edgecolors="black",
            s=x_axis, alpha=0.75)
plt.show()
```
