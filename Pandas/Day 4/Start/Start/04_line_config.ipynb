```python
# Dependencies
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Set x axis and variables
x_axis = np.arange(0, 10, 0.1)
sin = np.sin(x_axis)
cos = np.cos(x_axis)
```

* List of 'Markers' - https://matplotlib.org/api/markers_api.html
* List of 'Colors' - https://matplotlib.org/api/colors_api.html



```python
%matplotlib inline
plt.hlines(0,0,10,alpha=0.25)
```

```python
plt.hlines(3,5,10,alpha=0.25)
```

```python
plt.hlines(3,5,10,alpha=0.25)
plt.hlines(6,0,3,alpha=1.0)
```

```python
plt.hlines(3,5,10,alpha=0.25)
plt.hlines(6,0,3,alpha=1.0)
axes = plt.gca()
axes.set_xlim([0,4])
axes.set_ylim([0,10])
```

```python
sin_handle, = plt.plot(x_axis, sin, marker='o')
```

```python
sin_handle, = plt.plot(x_axis, sin, marker='o', label="Sin")
cos_handle, = plt.plot(x_axis, cos, marker='^', label="Cos")
```

```python
sin_handle, = plt.plot(x_axis, sin, marker='o', label="Sin")
cos_handle, = plt.plot(x_axis, cos, marker='^', label="Cos")
plt.legend(handles=[sin_handle, cos_handle], loc="lower right")
```
