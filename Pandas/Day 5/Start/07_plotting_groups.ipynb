```python
# Import Dependencies
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import os
```

```python
# Import our data into pandas from CSV
used_string = os.path.join("Resources", "UsedCars.csv")
used_car_df = pd.read_csv(used_string)

used_car_df
```

```python
used_car_df['maker'].value_counts()
```

## Plt Style examples:

Sometimes you want to tweak the default style a bit.

* http://www.futurile.net/2016/02/27/matplotlib-beautiful-plots-with-style/
* https://matplotlib.org/users/customizing.html

```python
pie_chart = used_car_df.maker.value_counts().plot(kind='pie', figsize=(8,8), autopct='%1.f%%')
plt.show()
```

```python
plt.style.use('fivethirtyeight')
plt.rcParams['xtick.labelsize'] = 10
plt.rcParams['font.size'] = 8
pie_chart = used_car_df.maker.value_counts().plot(kind='pie', figsize=(8,8), autopct='%1.f%%')
pie_chart.set_ylabel('')
plt.show()
```

```python
plt.style.use('ggplot')
plt.rcParams['xtick.labelsize'] = 10
plt.rcParams['font.size'] = 8
pie_chart = used_car_df.maker.value_counts().plot(kind='pie', figsize=(8,8), autopct='%1.f%%')
pie_chart.set_ylabel('')
plt.show()
```

```python
# Flush RCParams back to defaults
plt.rcdefaults()
```

```python
plt.style.use('seaborn')
plt.rcParams['xtick.labelsize'] = 10
plt.rcParams['font.size'] = 8
pie_chart = used_car_df.maker.value_counts().plot(kind='pie', figsize=(8,8), autopct='%1.f%%')
pie_chart.set_ylabel('')
plt.show()
```
