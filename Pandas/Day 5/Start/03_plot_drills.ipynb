```python
# Import Dependencies
import numpy as np
import matplotlib.pyplot as plt
```

```python
# DATASET 1
gyms = ["Crunch", "Planet Fitness", "NY Sports Club", "Rickie's Gym"]
members = [49, 92, 84, 53]
```

```python
x_axis = np.arange(0, len(gyms))
tick_locations = []
for x in x_axis:
    tick_locations.append(x)
    
plt.title("NYC Gym Popularity")
plt.ylabel("Number of Members")
plt.bar(x_axis, members, facecolor='red', alpha=.75)
plt.xticks(tick_locations, gyms)
plt.show()
```

```python
# DATASET 2
x_lim = 2 * np.pi
x_axis = np.arange(0, x_lim, 0.1)
sin = np.sin(x_axis)
```

```python
plt.hlines(0,0,x_lim, alpha=.2)
plt.plot(x_axis, sin, marker='o', color='r', linewidth=1)
plt.show()
```

```python
# DATASET 3
gyms = ["Crunch", "Planet Fitness", "NY Sports Club", "Rickie's Gym"]
members = [49, 92, 84, 53]
x_axis = np.arange(0, len(gyms))
colors = ["yellowgreen", "red", "lightcoral", "lightskyblue"]
explode = (0, 0.05, 0, 0)
```

```python
plt.title("NYC Gym Popularity")
plt.pie(members, explode=explode, labels=gyms, colors=colors, autopct="%1.1F%%", shadow=True)
plt.axis('equal')
plt.show()
```
