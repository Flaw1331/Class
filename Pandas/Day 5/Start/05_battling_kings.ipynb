```python
# Dependencies
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# Read CSV
got_data = pd.read_csv("Resources/got.csv")
got_data
```

```python
# Get attacker and defender data
attacker_data = got_data["attacker_king"].value_counts()
defender_data = got_data["defender_king"].value_counts()
defender_data
```

```python
# Since kings are sometimes attackers and sometimes defenders we need
# to ADD the two dataframes together to get their total battles
# Get total battle data
```

```python
combined_data = defender_data.add(attacker_data, fill_value=0)
combined_data.head()
```

```python
# Use pandas to plot a bar chart. Use figsize=(14,3) and rot=0 to match chart below
battle_plot = combined_data.plot(kind='bar', figsize=(14,3), rot=0)
```

```python
# Set textual properties
plt.title("The Bloodthirst of Kings")
plt.ylabel("Number of Battles Participated In")
```
