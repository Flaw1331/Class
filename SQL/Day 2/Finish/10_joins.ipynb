* Notice that the loser_id, winner_id and the player_id all refer to the same id

* The **`INNER JOIN`** Returns records that have matching values in both tables.

```python
sql_query = """
select *
from matches
inner join player_roster
on matches.winner_id = player_roster.player_id;
"""

inner_join = pd.read_sql_query(sql_query, engine)
inner_join
```

```python
sql_query = """
select first_name, last_name, tourney_name, score
from matches
inner join player_roster
on matches.winner_id = player_roster.player_id;
"""

inner_join = pd.read_sql_query(sql_query, engine)
inner_join.head()

```

```python
sql_query = """
select first_name, last_name, tourney_name, score
from matches
inner join player_roster
on matches.winner_id = player_roster.player_id
where winner_id = 200003;
"""

inner_join = pd.read_sql_query(sql_query, engine)
inner_join
```
