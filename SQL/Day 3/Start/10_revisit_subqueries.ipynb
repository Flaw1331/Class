```python
sql_query = """
    select count(*)
    from customer
    where customer_id in
    (
        select customer_id
        from payment
        where rental_id in
        (
            select rental_id
            from rental
            where inventory_id in
            (
                select inventory_id
                from inventory
                where film_id in
                (
                    select film_id
                    from film
                    where title = 'Blanket Beverly'
                )
            )
        )
    );
"""
film = pd.read_sql_query(sql_query, engine)
film.head()
```

```python
sql_query = '''
select distinct count(*) as numberofpeople
from film f
inner join inventory i on f.film_id = i.film_id
inner join rental r on r.inventory_id = i.inventory_id
inner join customer c on c.customer_id = r.customer_id
where title = 'Blanket Beverly'

'''
films = pd.read_sql_query(sql_query,engine)
films
```
