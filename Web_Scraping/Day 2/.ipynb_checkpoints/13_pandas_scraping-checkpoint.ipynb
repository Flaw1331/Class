```python
import pandas as pd
```

We can use the read_html function in Pandas to automatically scrape any tabular data from a page.

```python
url = 'https://en.wikipedia.org/wiki/List_of_capitals_in_the_United_States'
```

```python
tables = pd.read_html(url)
tables
```

* instead of a dataframe, what we get in return is a list of dataframes for any tabular data that Pandas found.

```python
type(tables)
```

We can slice off any of those dataframes that we want using normal indexing.

```python
df = tables[0]
df
```

```python
df.columns = ['State', 'Abr.', 'State-hood Rank', 'Capital', 
              'Capital Since', 'Area (sq-mi)', 'Municipal Population', 
              'Metropolitan Population', 'Rank in state', 'Rank in US',  'Notes']
```

```python
df
```

```python
df = df.iloc[2:]
df.head()
```

Set the index to the State column

```python
df.set_index('State', inplace=True)
df.head()
```

```python
df.loc['California']
```

### Dataframes to HTML

* Pandas also has a `to_html` method that we can use to generate HTML tables from DataFrames.

```python
html_table = df.to_html()
html_table
```

You may have to strip unwanted newlines to clean up the table.

```python
html_table.replace('\n', '')
```

You can also save the table directly to a file.

```python
df.to_html('table.html')
```
