```python
# Dependencies
from bs4 import BeautifulSoup
import requests

# URL of page to be scraped
url = 'https://kansascity.craigslist.org/search/sss?sort=rel&query=guitar'
```

```python
# Retrieve page with the requests module
response = requests.get(url)
# Create BeautifulSoup object; parse with 'html.parser'
soup = BeautifulSoup(response.text, 'html.parser')

# Examine the results, then determine element that contains sought info
print(soup.prettify())
```

```python
with open("craigslist.html", "w") as file:
    file.write(response.text)
```

To find results, it's often easier to use Visual Studio Code than scrolling through IPython Notebook windows.

In Visual Studio Code you can use (Apple-K, Apple-F) or (Command-K, Command-F) to format your html file. This will help clean things up. Use the collapse icons on the left hand gutter to find the chuck of HTML you need to process.

```python
# results are returned as an iterable list
results = soup.find_all('li', class_="result-row")
results
```

```python
for result in results:
    try:
        title = result.find('a', class_="result-title").text
        price = result.a.span.text
        link = result.a['href']

        if (title and price and link):
            print('-------------')
            print(f'{price}. {title}')
            print(link)
    except Exception as e:
        print(e)


```
