```python
import json
import requests
url = "https://api.nytimes.com/svc/search/v2/articlesearch.json?"
api_key = "164b73c522a8420c8e05343ef1da0a7e"
```

```python
# Store a search term
q = "obama"
```

```python
# Search for articles published between a begin and end date
begin_date = "20160101"
end_date = "20160130"
```

```python
query = url + "api-key=" + api_key + "&q=" + q + "&begin_date=" + begin_date\
    + "&end_date=" + end_date
```

```python
articles = requests.get(query).json()
print(articles)

#article_list = [article for article in articles["response"]["docs"]]
```

```python
article_list = [article for article in articles["response"]["docs"]]
```

```python
for article in article_list:
    print("A snippet from the article: '" + article["snippet"] + "'.\n")
```

```python
article_list = []
for page in range(0, 3):
    page_query = query + "&page=" + str(page)
    articles = requests.get(page_query).json()
    print(page_query)
    article_list.extend([article for article in articles["response"]["docs"]])
```
