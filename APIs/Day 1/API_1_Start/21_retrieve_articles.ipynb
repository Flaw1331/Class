```python
import json
import requests as req
import time
url = "https://api.nytimes.com/svc/search/v2/articlesearch.json?"
api_key = "1ecabb81c59f42619112b9611cf9949c"
```

```python
q = "Obama"
begin_date = "20171230"
end_date = "20171231"
```

```python
query = url + "api-key=" + api_key + "&q=" + q + "&begin_date" + begin_date + "&end_date" + end_date
print(query)
```

```python
articles = req.get(query).json()
```

```python
article_list = [article for article in articles['response']['docs']]
```

```python
article_list = []
for page in range(0,3):
    page_query = query + "&page=" + str(page)
    articles = req.get(page_query).json()
    print(page_query)
    time.sleep(1)
    article_list.extend([article for article in articles['response']['docs']])
```
